```python
#A) Convert runtime from string to integer
df["Runtime (Minutes)"] = df["Runtime (Minutes)"].astype(str).str.replace(" min", "", regex=False).astype(int)
df["Runtime (Minutes)"].head()
```

```python
#B) Group by certificate and calculate mean, median, and std of  => Group by director and calculate mean, median, and std of 
df.groupby("Director")["Rating"].agg(["mean", "median", "std"])
```

```python
#C) Sort by rating and votes descending
df.sort_values(by=["Rating", "Votes"], ascending=[False, False])
```

```python
#D) Float gross and replace missing with median
df["Revenue (Millions)"] = df["Revenue (Millions)"].str.replace(",", "", regex=False).str.replace("$", "", regex=False)
df["Revenue (Millions)"] = pd.to_numeric(df["Gross"], errors="coerce")
gross_median = df["Revenue (Millions)"].median()
df["Revenue (Millions)"] = df["Revenue (Millions)"].fillna(gross_median)
df["Revenue (Millions)"].head()
```

```python
#E) Movie length
bins = [0, 90, 150, 500]
labels = ["Short", "Standard", "Epic"]

df["Movie_Length"] = pd.cut(df["Runtime (Minutes)"], bins=bins, labels=labels)
df["Movie_Length"]
```

```python
#F) Filter
genres_list = ["Drama", "Action", "Comedy"]
df[df["Genre"].isin(genres_list)]
```

```python
#G) Highest-rated movie for every year
idx = df.groupby("Year")["Rating"].idxmax()
df.loc[idx]
```

```python
#H) Split and explode genre
df["Genre"] = df["Genre"].str.split(", ")
df = df.explode("Genre")
df["Genre"].head()
```

```python
#I) Query for highly rated and voted movies
df.query("Rating > 8.0 and Votes > 100000")
```

```python
#K) Top 5 longest R-rated movies
df[df["Director"] == "R"].nlargest(5, "Runtime (Minutes)")
```
